# 04 — Fine-Tuning

**Project:** AuraVision — Crowd Age Group Majority Classifier  
**Purpose:** Fine-tune the base model with stronger regularisation to reduce overfitting.

**Key differences from initial training:**
- Backbone fully frozen — only the classifier head is retrained.
- Heavier dropout (0.5 / 0.4) and stronger weight decay (1e-2).
- Cosine annealing with warm restarts.
- Test-time augmentation (TTA) for the final accuracy estimate.
